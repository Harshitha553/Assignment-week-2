### SQL CASE WHEN Statement

The CASE WHEN statement is used to add IF-ELSE logic in SQL.

Syntax:

SELECT column_name,
       CASE
           WHEN condition1 THEN result1
           WHEN condition2 THEN result2
           ELSE result
       END
FROM table_name;

Example Table: Employees

| EmpID | Name      | Salary |
| ----- | --------- | ------ |
| 1     | Harshitha | 30000  |
| 2     | Bhargav   | 60000  |
| 3     | Riya      | 90000  |


1. Classify employees by salary
SELECT Name,
       Salary,
       CASE
           WHEN Salary < 50000 THEN 'Low Salary'
           ELSE 'High Salary'
       END AS Salary_Category
FROM Employees;

## Output:

| Name      | Salary | Salary_Category |
| --------- | ------ | --------------- |
| Harshitha | 30000  | Low Salary      |
| Bhargav   | 60000  | High Salary     |
| Riya      | 90000  | High Salary     |

2. Multiple Conditions

SELECT Name,
       Salary,
       CASE
           WHEN Salary < 40000 THEN 'Low'
           WHEN Salary BETWEEN 40000 AND 80000 THEN 'Medium'
           ELSE 'High'
       END AS Salary_Band
FROM Employees;  

3. CASE in ORDER BY

Sort females first.

SELECT *
FROM Employees
ORDER BY
CASE
    WHEN Gender='Female' THEN 1
    ELSE 2
END;

4. CASE with Aggregate Functions

Count Male and Female employees.

SELECT
COUNT(
CASE
    WHEN Gender='Male'
    THEN 1
END
) AS Male_Count,

COUNT(
CASE
    WHEN Gender='Female'
    THEN 1
END
) AS Female_Count

FROM Employees;

Output:

| Male_Count | Female_Count |
| ---------- | ------------ |
| 5          | 4            |


### What happens if no condition is true in a CASE expression and there is no ELSE clause?

It returns NULL.

### ACID Properties

ACID properties ensure that database transactions are processed reliably and safely.

ACID = Atomicity + Consistency + Isolation + Durability

## What is a Transaction?

A transaction is a group of SQL operations executed as a single unit.

BEGIN TRANSACTION;

UPDATE Accounts
SET Balance = Balance - 1000
WHERE AccountID = 1;

UPDATE Accounts
SET Balance = Balance + 1000
WHERE AccountID = 2;

COMMIT;

1. Atomicity (All or Nothing)

Atomicity ensures a transaction is executed completely or not executed at all.

2. Consistency
Rule

Database must always move from one valid state to another valid state.

Example

Before transaction:

A = 5000
B = 3000

Total = 8000

After transfer:

A = 4000
B = 4000

Total = 8000

Consistency maintained ✅

3. Isolation
Rule

Multiple transactions should not interfere with each other.

4. Durability
Rule

Once a transaction is committed, data remains permanently stored.

Example
COMMIT;

Suppose power failure occurs immediately after commit.

Data should still exist after restart.
