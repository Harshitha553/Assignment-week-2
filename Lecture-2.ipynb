SQL WHERE Conditions

The WHERE clause is used to filter records based on conditions.

SELECT * FROM table_name WHERE condition;

1. AND Operator

Returns rows when all conditions are true.

SELECT * FROM Students WHERE Marks > 80 AND City='Hyderabad';

2. OR Operator

Returns rows when at least one condition is true.

SELECT * FROM Students WHERE City='Hyderabad' OR Marks > 90;

3. NOT Operator

Returns rows that do NOT satisfy a condition.

SELECT *
FROM Students
WHERE NOT City='Hyderabad';

4. BETWEEN Operator

Used to select values within a range.

SELECT *
FROM Students
WHERE Marks BETWEEN 80 AND 90;

5. IN Operator

Used when checking multiple values.

Instead of:

SELECT *
FROM Students
WHERE City='Hyderabad'
OR City='Mumbai';

Use:

SELECT *
FROM Students
WHERE City IN ('Hyderabad','Mumbai');

6. NOT IN
SELECT *
FROM Students
WHERE City NOT IN ('Hyderabad','Mumbai');

7. LIKE Operator

Used for pattern matching.

Starts with H
SELECT *
FROM Students
WHERE Name LIKE 'H%';

Output:

Harshitha
Ends with a
SELECT *
FROM Students
WHERE Name LIKE '%a';

Output:

Riya
Contains ar
SELECT *
FROM Students
WHERE Name LIKE '%ar%';

Output:

Harshitha

8. Wildcards
Symbol	Meaning
%	Any number of characters
_	Single character

Example:

SELECT *
FROM Students
WHERE Name LIKE '_iya';

Matches:

Riya
