### SQL GROUP BY

GROUP BY is used to group rows that have the same values in a column and perform aggregate functions on each group.

1. Count Employees by Gender
SELECT Gender, COUNT(*) AS TotalEmployees
FROM Employee
GROUP BY Gender;

Output:

Gender	TotalEmployees
Male	3
Female	2

2. Sum Salary by Gender
SELECT Gender,
       SUM(Salary) AS TotalSalary
FROM Employee
GROUP BY Gender;



### GROUP BY with WHERE

Find employees earning more than 5000.

SELECT Gender,
       COUNT(*)
FROM Employee
WHERE Salary > 5000
GROUP BY Gender;

### HAVING Clause
Used to filter groups.

SELECT Gender,
       COUNT(*) AS Total
FROM Employee
GROUP BY Gender
HAVING COUNT(*) > 2;

Difference Between WHERE and HAVING
WHERE	               HAVING
Filters rows	     Filters groups
Before GROUP BY	     After GROUP BY

WHERE  → Filters Rows
GROUP BY → Creates Groups
HAVING → Filters Groups
