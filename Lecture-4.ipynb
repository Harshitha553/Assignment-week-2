## SQL Joins

Joins are used to combine data from multiple tables using a related column.

### Usage of Joins

1. Recombine data
2. Data Enrichment "Getting Extra Data" - You are working on one table but want to retrieve info from another table we use joins.
3. Check for Existence "Filtering"

Example Table-1 (Employees)

| EmpID | Name      | DeptID |
| ----- | --------- | ------ |
| 1     | Harshitha | 101    |
| 2     | Bhargav   | 102    |
| 3     | Riya      | 103    |
| 4     | Rahul     | NULL   |

Example Table-2 (Departments)

| DeptID | DeptName |
| ------ | -------- |
| 101    | CSE      |
| 102    | ECE      |
| 104    | MECH     |


1. INNER JOIN

Returns only matching rows from both tables.

SELECT e.Name,d.DeptName
FROM Employees e
INNER JOIN Departments d
ON e.DeptID = d.DeptID;

Output:

| Name      | DeptName |
| --------- | -------- |
| Harshitha | CSE      |
| Bhargav   | ECE      |


2. LEFT JOIN

Returns:

All rows from LEFT table
Matching rows from RIGHT table

SELECT e.Name,d.DeptName
FROM Employees e
LEFT JOIN Departments d
ON e.DeptID = d.DeptID;

Output:

| Name      | DeptName |
| --------- | -------- |
| Harshitha | CSE      |
| Bhargav   | ECE      |
| Riya      | NULL     |
| Rahul     | NULL     |


3. RIGHT JOIN

Returns:

All rows from RIGHT table
Matching rows from LEFT table

SELECT e.Name,d.DeptName
FROM Employees e
RIGHT JOIN Departments d
ON e.DeptID = d.DeptID;

Output:

| Name      | DeptName |
| --------- | -------- |
| Harshitha | CSE      |
| Bhargav   | ECE      |
| NULL      | MECH     |


4. FULL OUTER JOIN

Returns:

All rows from left table
All rows from right table
Matching wherever possible

SELECT e.Name,d.DeptName
FROM Employees e
FULL OUTER JOIN Departments d
ON e.DeptID = d.DeptID;

Output:

| Name      | DeptName |
| --------- | -------- |
| Harshitha | CSE      |
| Bhargav   | ECE      |
| Riya      | NULL     |
| Rahul     | NULL     |
| NULL      | MECH     |


Create these tables and practice:

CREATE TABLE Employees(
EmpID INT,
Name VARCHAR(50),
DeptID INT,
Salary INT
);

CREATE TABLE Departments(
DeptID INT,
Department VARCHAR(50)
);

Q1

Display employee name and department name using JOIN.

SELECT e.Name, d.Department
FROM Employees e
INNER JOIN Departments d
ON e.DeptID = d.DeptID;

Q2

Find total salary department-wise.

SELECT d.Department,
       SUM(e.Salary) AS TotalSalary
FROM Employees e
INNER JOIN Departments d
ON e.DeptID = d.DeptID
GROUP BY d.Department;

Q3

Find average salary department-wise.

SELECT d.Department,
       AVG(e.Salary) AS AvgSalary
FROM Employees e
INNER JOIN Departments d
ON e.DeptID = d.DeptID
GROUP BY d.Department;

Q4

Find departments having more than 5 employees.

(Hint: GROUP BY + HAVING)

SELECT d.Department,
       COUNT(*) AS EmployeeCount
FROM Employees e
INNER JOIN Departments d
ON e.DeptID = d.DeptID
GROUP BY d.Department
HAVING COUNT(*) > 5;

Q5

Find customer who spent the highest amount.

SELECT TOP 1
       Customer,
       SUM(Amount) AS TotalSpent
FROM Orders
GROUP BY Customer
ORDER BY TotalSpent DESC;
