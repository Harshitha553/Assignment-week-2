What is SQL?

SQL (Structured Query Language) is used to:

Store data
Retrieve data
Update data
Delete data

1. SELECT

Used to retrieve data.

SELECT * FROM Employee;

Output:
All rows and columns.

Specific columns:

SELECT Name, Salary
FROM Employee;

2. WHERE

Used to filter records.

Employees from IT:

SELECT *
FROM Employee
WHERE Department = 'IT';

3. ORDER BY

Sort data.

Salary ascending:

SELECT *
FROM Employee
ORDER BY Salary;

Salary descending:

SELECT *
FROM Employee
ORDER BY Salary DESC;

4. DISTINCT

Remove duplicates.
SELECT DISTINCT Department
FROM Employee;

5. LIMIT

Get top rows.

SELECT *
FROM Employee
LIMIT 2;

SELECT TOP 2 FROM Employee;

What is GROUP BY?

GROUP BY creates groups of rows having the same value.

For example:

Departments:

IT
HR
IT
Sales
HR

GROUP BY Department creates:

IT Group
HR Group
Sales Group

1. Count Employees in Each Department
SELECT Department, COUNT(*) AS EmployeeCount
FROM Employee
GROUP BY Department;

2. Average Salary by Department
SELECT Department, AVG(Salary) AS AvgSalary
FROM Employee
GROUP BY Department;

HAVING

WHERE filters rows before grouping.

HAVING filters groups after grouping.

Q. Show departments having more than 1 employee.

SELECT Department, COUNT(*) AS EmployeeCount
FROM Employee
GROUP BY Department
HAVING COUNT(*) > 1;

Q. Departments whose average salary is above 50000.

SELECT Department, AVG(Salary) AS AvgSalary
FROM Employee
GROUP BY Department
HAVING AVG(Salary) > 50000;

### 1. Primary Key

A Primary Key is a column (or combination of columns) that uniquely identifies each row in a table.

Properties

✅ Unique values

✅ Cannot be NULL

✅ One primary key per table

CREATE TABLE Students (
    StudentID INT PRIMARY KEY,
    Name VARCHAR(50),
    Age INT
);

### 2. Constraints 

Constraints are rules applied on table columns to maintain data accuracy and integrity.

1. NOT NULL

Column cannot contain NULL values.

CREATE TABLE Students (
    StudentID INT,
    Name VARCHAR(50) NOT NULL
);

2. UNIQUE

Ensures all values are different.

CREATE TABLE Students (
    Email VARCHAR(100) UNIQUE
);

3. PRIMARY KEY
StudentID INT PRIMARY KEY

Combination of:

UNIQUE + NOT NULL

4. FOREIGN KEY

Creates relationship between tables.

CREATE TABLE Students (
    StudentID INT PRIMARY KEY,
    DepartmentID INT,
    FOREIGN KEY (DepartmentID)
    REFERENCES Departments(DepartmentID)
);

5. CHECK

Validates condition before inserting data.

CREATE TABLE Students (
    Age INT CHECK (Age >= 18)
);

6. DEFAULT

Assigns default value.

CREATE TABLE Students (
    City VARCHAR(50) DEFAULT 'Hyderabad'
);

If city not provided:

Hyderabad

is automatically inserted.
