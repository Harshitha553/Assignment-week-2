## Step-6 Solve use cases 

### 1. Monthly Trend

### select year(str_to_date(`Order Date`,'%m/%d/%Y')) as year_,month(str_to_date(`Order Date`,'%m/%d/%Y')) as month_, sum(sales) as total_sales from `sample - superstore` group by year_,month_ order by year_,month_;



#### Insight:
#### The query analyzes sales trends over time and helps identify monthly variations in revenue.

### 2. Top 10 customers

### select `Customer Name`,sum(sales) as total_sales from `sample - superstore` group by `Customer Name` order by total_sales desc limit 10;



#### Insight:
#### The query identifies customers who contributed the most revenue.

### 3. Duplicate orders

### select `Order ID`,count(*) as duplicate_count from `sample - superstore` group by `Order ID` having count(*)>1;



#### Insight:
#### The query checks for repeated Order IDs and helps assess data consistency.
